<h2>3. 👀Data Visualization</h2>

```python
plt.figure(figsize=(4, 4))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
sns.lmplot(x='Study Hours', y="Grades", data=df, height=4, aspect=1.5,)
plt.title("Study Hours vs. Grades")
sns.lmplot(x='Attendance (%)', y="Grades", data=df, height=4, aspect=1.5,)
plt.title("Attendance (%) vs. Grades")
sns.lmplot(x='Socioeconomic Score', y="Grades", data=df, height=4, aspect=1.5,)
plt.title("Socioeconomic Score vs. Grades")
sns.lmplot(x='Sleep Hours', y="Grades", data=df, height=4, aspect=1.5,)
plt.title("Sleep Hours Score vs. Grades")
plt.tight_layout()
plt.show()
```

<h2>4. 🗒️Conclusion</h2>

<p>As we can see the feature that has most correlation to the Grade is Study Hours, while another features like Socioeconomic Score, and Attendance play a little role to the Grades, it still probably good if we want to use those features in the model to predict the Grades. And now we know that Sleep Hours is doesn't affect the Grades at least in terms of Linear.</p>
